```python
y=data['diagnosis']
X=data.drop('diagnosis',axis=1)
X.shape,y.shape
```

##Normalize the Data

```python
from sklearn.preprocessing import StandardScaler

#Create a scalar object
scaler=StandardScaler()

# fit the scalar to the data and transform the data
X_scaled=scaler.fit_transform(X)

X_scaled
```

```python
import torch
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,
                                               y,
                                               test_size=0.3,
                                               random_state=42) #want to repeat the split this is number to pass

X_train=torch.from_numpy(X_train).float()
y_train=torch.from_numpy(y_train.values).float()
X_test=torch.from_numpy(X_test).float()
y_test=torch.from_numpy(y_test.values).float()

X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
import torch
import torch.nn as nn
class LogisticRegressionModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.weights=nn.Parameter(torch.randn((30,1),requires_grad=True))
    self.bias=nn.Parameter(torch.randn(1,requires_grad=True))
  def forward(self,x):
    y=torch.matmul(x,self.weights)+self.bias
    return y
```

```python
model_0=LogisticRegressionModel()
list(model_0.parameters())
```

```python
try:
  from helper_functions import accuracy_fn
except:
  print('[INFO] counld not find the module downloading the required files...')
  import requests
  with open('helper_functions.py','wb') as f:
    response=requests.get('https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/helper_functions.py')
    f.write(response.content)
  from helper_functions import accuracy_fn
```

```python
#optimizer and loss
loss_fn=nn.BCELoss()
optimizer=torch.optim.SGD(params=model_0.parameters(),lr=0.1)
```

```python
type(X_train)
```

```python
#creating the training loop
train_loss=[]
test_loss=[]
train_acc=[]
test_acc=[]
epochs=100
for epoch in range(epochs):
  model_0.train()
  y_logits=model_0(X_train)
  y_probs=torch.sigmoid(y_logits)
  y_preds=(y_probs>=0.5).int()

  loss=loss_fn(y_probs.squeeze(),y_train.float())
  acc=accuracy_fn(y_true=y_train.float(),y_pred=y_preds.squeeze())

  train_loss.append(loss.item())
  train_acc.append(acc)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  model_0.eval()
  with torch.inference_mode():
    test_logits=model_0(X_test)
    test_probs=torch.sigmoid(test_logits)
    test_preds=(test_probs>=0.5).int()
    # print(test.dtype)
    loss_=loss_fn(test_probs.squeeze(),y_test)
    acc_=accuracy_fn(y_true=y_test.float(),y_pred=test_preds.squeeze())

    test_loss.append(loss_.item())
    test_acc.append(acc_)

  if epoch %10 ==0:
    print(f'Loss: {loss:.2f} | Accuracy: {acc:.2f}% | Test loss: {loss_:.2f} | Test accuracy: {acc_:.2f}%')

```

```python
import matplotlib.pyplot as plt
# Plotting
plt.figure(figsize=(12, 5))

# Plotting Losses
plt.subplot(1, 2, 1)
plt.plot(train_loss, label='Train Loss')
plt.plot(test_loss, label='Test Loss')
plt.title('Loss Curves')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plotting Accuracies
plt.subplot(1, 2, 2)
plt.plot(train_acc, label='Train Accuracy')
plt.plot(test_acc, label='Test Accuracy')
plt.title('Accuracy Curves')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
